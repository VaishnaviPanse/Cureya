**3.Define : Dictionaries, lists, tuples, set with an example**


**Dictionary:**

It is data type which is use to store data in key:value pairs. It is orderd, changable and does not allow duplicate values.

```python
#empty dict
a={}
print (a)

#dictionary containing name of some countries and their capital
b={
    "India": "Delhi",
   "Japan": "Tokeyo",
   "Egypt": "Cairo",
   "France":"Paris"
}
print(b)
```

**List**:
 
List is a data type use to store collection of data. It is orderd changable and allows duplicate values. 

```python
#Empty list
a=[]
print(a)

b=[3,"Mango",9.7,True] #It can contain values of different data types 
print(b)
```

**Tuples**: 

Tuple is a built in data type in python which is use to store store collecton of data. It is orederd and unchangable.

```python
#empty tuple
a=()
print(a)

b=(4,5,"Apple")
print(b)
```

**Sets**: 

Set is a built in data type in python which is use to store collection of data. It is unorderd and unindexed. 

```python
#Examples
a={} #empty set
print(a)

b={1,2,3,"Mango",False}
print(b)
```

2. Read all product sales data and show it using a multi line plot. Display the number of
    units sold per month for each product using multi line plots. (i.e., Separate Plotline for
    each product ).

```python
months  = data ['month_number'].tolist()
faceCream_SalesData = data ['facecream'].tolist()
faceWash_SalesData  = data ['facewash'].tolist()
toothPaste_SalesData = data ['toothpaste'].tolist()
bathingsoap_SalesData   = data ['bathingsoap'].tolist()
shampoo_SalesData   = data ['shampoo'].tolist()
moisturizer_SalesData = data ['moisturizer'].tolist()

plt.plot(months, faceCream_SalesData,   label = 'Facecream Sales Data', marker='o', linewidth=2)
plt.plot(months, faceWash_SalesData,   label = 'FaceWash Sales Data',  marker='o', linewidth=2)
plt.plot(months, toothPaste_SalesData, label = 'ToothPaste Sales Data', marker='o', linewidth=2)
plt.plot(months, bathingsoap_SalesData, label = 'BathingSoap Sales Data', marker='o', linewidth=2)
plt.plot(months, shampoo_SalesData, label = 'Shampoo Sales Data', marker='o', linewidth=2)
plt.plot(months, moisturizer_SalesData, label = 'Moisturizer Sales Data', marker='o', linewidth=2)

plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.legend(loc='upper left')
plt.xticks(months)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
```
